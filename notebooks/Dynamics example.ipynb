```python
import dynamicalab as dlb
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

#### Generate graph

```python
G = nx.erdos_renyi_graph(20,0.5)
T = np.linspace(0,1,200)
```

### SIS dynamics

```python
from dynamicalab.dynamics import SISDynamics
```

```python
model = SISDynamics(0.2,0.3)
```

```python
X = model(G, T)
```

```python
plt.figure(dpi=150)
plt.imshow(X.T)
plt.xlabel("Time"), plt.ylabel("Node")
```

## Theta model

```python
from dynamicalab.dynamics import ThetaModelDynamics
```

```python
model = ThetaModelDynamics(sigma=10, input_intensity=1e-5)
```

```python
T = np.linspace(0,10,500)
```

```python
X = model(G, T)
```

## Bernoulli dynamics

```python
from dynamicalab.dynamics import BernoulliDynamics
```
